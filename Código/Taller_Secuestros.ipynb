```python
import sqlite3
import pandas as pd
```

```python
DF = pd.read_json("https://www.datos.gov.co/resource/d7zw-hpf4.json?$limit=100000")
DF
```

```python
conn = sqlite3.connect('Secuestros.db')
cursor = conn.cursor()
DF.to_sql('secuestros', conn, if_exists='replace', index=False)
def ejecutar_sql(query, conexion=conn):
  """
  Toma una consulta SQL, la ejecuta y devuelve el resultado
  en una tabla de pandas para una visualización clara.
  """
  df = pd.read_sql_query(query, conexion)
  # Usamos display() porque en Colab presenta las tablas de forma más elegante.
  display(df)
```

```python
ejecutar_sql("SELECT DISTINCT departamento FROM secuestros ORDER BY departamento;")
```

```python
ejecutar_sql("""
SELECT 
    departamento,
    tipo_delito,
    COUNT(*) AS Total
FROM 
    secuestros
GROUP BY 
    departamento,
    tipo_delito

""")
```

```python
ejecutar_sql("""
SELECT 
    SUBSTR(fecha_hecho, 1, 4) AS año,
    SUM(cantidad) AS total_delitos
FROM 
    secuestros
GROUP BY 
    año
ORDER BY 
    año;
""")
```

```python
ejecutar_sql("""
SELECT 
    tipo_delito,
    SUM(cantidad) AS total
FROM 
    secuestros
WHERE 
    fecha_hecho LIKE '2020%'
GROUP BY 
    tipo_delito
ORDER BY 
    total DESC;
""")
```

```python
# Cerramos la conexión
conn.close()
print("Conexión a la base de datos cerrada.")
```

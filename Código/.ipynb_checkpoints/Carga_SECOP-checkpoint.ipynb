```python
import pandas as pd
import os
from glob import glob
ruta = "./datos_secop"
archivos_sin_b = sorted([f for f in glob(os.path.join(ruta, "*.csv")) if "B" not in f])
archivos_b2 = sorted(glob(os.path.join(ruta, "*B2.csv")))
archivos_b3 = sorted(glob(os.path.join(ruta, "*B3.csv")))
archivos_b4 = sorted(glob(os.path.join(ruta, "*B4.csv")))
archivos_b5 = sorted(glob(os.path.join(ruta, "*B5.csv")))
df_b1 = pd.concat([pd.read_csv(archivo) for archivo in archivos_sin_b], ignore_index=True)
df_b2 = pd.concat([pd.read_csv(archivo) for archivo in archivos_b2], ignore_index=True)
df_b3 = pd.concat([pd.read_csv(archivo) for archivo in archivos_b3], ignore_index=True)
df_b4 = pd.concat([pd.read_csv(archivo) for archivo in archivos_b4], ignore_index=True)
df_b5 = pd.concat([pd.read_csv(archivo) for archivo in archivos_b5], ignore_index=True)
```

```python
df_final = pd.concat([df_b1, df_b2, df_b3,df_b4,df_b5], axis=1)
```

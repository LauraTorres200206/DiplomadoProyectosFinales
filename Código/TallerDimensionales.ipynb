```python
ColumnsImputar = [
    'desercion_transicion', 'desercion_media', 'desercion_secundaria',
    'desercion_primaria', 'repitencia_transicion', 'repitencia_secundaria',
    'reprobacion_media', 'repitencia_primaria', 'repitencia', 'repitencia_media',
    'desercion', 'reprobacion_secundaria', 'cobertura_bruta_media',
    'tasa_matriculacion_5_16', 'cobertura_neta', 'cobertura_bruta_transicion',
    'reprobacion_primaria', 'cobertura_bruta_primaria', 'cobertura_neta_primaria',
    'reprobacion', 'aprobacion_transicion', 'reprobacion_transicion',
    'aprobacion_media', 'cobertura_bruta_secundaria', 'cobertura_neta_secundaria',
    'cobertura_neta_media', 'cobertura_bruta', 'cobertura_neta_transicion',
    'aprobacion_secundaria'
]

df_imputado = df_educacion.copy()

for col in ColumnsImputar:
    df_imputado[col] = df_imputado.groupby('codigo_departamento')[col].transform(
        lambda x: x.fillna(x.median(skipna=True))
    )
    
```

```python

comparacion_medianas = pd.DataFrame({
    "Columna": ColumnsImputar,
    "Mediana Original (con nulos)": [df_educacion[col].median(skipna=True) for col in ColumnsImputar],
    "Mediana Imputada": [df_imputado[col].median() for col in ColumnsImputar]
})

comparacion_medianas["Diferencia absoluta"] = (
    comparacion_medianas["Mediana Imputada"] - comparacion_medianas["Mediana Original (con nulos)"]
)

comparacion_medianas["Diferencia relativa (%)"] = (
    100 * comparacion_medianas["Diferencia absoluta"] / (comparacion_medianas["Mediana Original (con nulos)"] + 1e-5)
).round(2)

comparacion_medianas
```

# Respecto a la población del municipio ¿Que porcentaje de escolaridad hay?

```python
consulta_1 = """
SELECT 
    dm.municipio,
    AVG(f.tasa_matriculacion_5_16) AS Promedio_escolaridad
FROM 
    hechos f
JOIN 
    Dimmunicipio dm ON f.id_municipio = dm.id_municipio
WHERE 
    dm.municipio <> 'NACIONAL'
GROUP BY 
    dm.municipio
ORDER BY 
    Promedio_escolaridad DESC;
"""
ejecutar_sql(consulta_1)
```

Los hallazgos que encontramos del porcentaje de escolaridad de la población entre 5 y 16 años es las diferencias territoriales en el acceso al sistema educativo. Municipios como Cota (209.95%) y Tenjo (149.60%) presentan tasas superiores al 100%, lo que puede atribuirse segun la informacion de Datos Abiertos a flujos migratorios internos no capturados por las proyecciones del DANE o a su rol como centros educativos regionales. En contraste, territorios como Papunaua (0.0%) y Pana Pana (27.5%) reflejan coberturas extremadamente bajas, lo que sugiere barreras estructurales de acceso, posibles vacíos institucionales o subregistro de datos. Estas variaciones resaltan la necesidad de mejorar estos porcentajes para garantizar el derecho a la educación de manera equitativa.

# ¿Cómo compararía el rendimiento educativo por municipios?

```python
consulta_2 = """
SELECT 
    dm.municipio,
    AVG(f.aprobaci_n) AS promedio_aprobacion,
    AVG(f.reprobacion) AS promedio_reprobacion,
    AVG(f.repitencia) AS promedio_repitencia
FROM 
    hechos f
JOIN 
    Dimmunicipio dm ON f.id_municipio = dm.id_municipio
WHERE 
    dm.municipio <> 'NACIONAL'
GROUP BY 
    dm.municipio
ORDER BY 
    promedio_aprobacion DESC;
"""
ejecutar_sql(consulta_2)
```

La forma en que se comparó el rendimiento educativo entre municipios fue a través de tres indicadores: la tasa de aprobación, reprobación y repitencia. Estos nos dan una idea general de cómo les está yendo a los estudiantes en cada uno de los municipios, si están logrando avanzar en su proceso educativo o si, por el contrario, están enfrentando barreras que los hacen repetir o reprobar.

Por ejemplo, municipios como Puerto Arica y Tibirita presentan tasas de aprobación muy altas, lo cual puede reflejar un buen acompañamiento académico, condiciones favorables para el aprendizaje o incluso una gestión educativa más sólida. En cambio, en lugares como Mapiripana o Cacahual, las tasas de reprobación y repitencia son más altas, lo que podría estar relacionado con problemas estructurales, falta de acceso a recursos educativos o dificultades en la permanencia escolar.

# ¿Que departamentos son los que mejor cobertura tienen? ¿Pueden hacer cálculo con SQL?

```python
consulta_3 = """
SELECT 
    dd.departamento,
    AVG(f.cobertura_neta) AS promedio_cobertura_neta,
    AVG(f.cobertura_bruta) AS promedio_cobertura_bruta
FROM 
    hechos f
JOIN 
    DimDepartamento dd ON f.id_Departamento = dd.id_Departamento
GROUP BY 
    dd.departamento
ORDER BY 
    promedio_cobertura_neta DESC;
"""
ejecutar_sql(consulta_3)
```

Para identificar qué departamentos tienen mejor cobertura educativa hacemos una consulta en SQL, en la que se revisaron dos indicadores clave: la cobertura neta (que mide cuántos niños y jóvenes entre 5 y 16 años están realmente asistiendo al sistema educativo) y la cobertura bruta (que incluye a toda la población matriculada, sin importar si están dentro del rango de edad esperado). A partir de esto, se promedió cada indicador por departamento y se ordenó de mayor a menor.

Al observar los datos, departamentos como Quindío, Sucre, Cesar y Magdalena se destacan por tener las coberturas netas más altas, lo que indica que la mayoría de los niños en edad escolar sí están estudiando. Además, sus coberturas brutas superan el 100%, lo que se puede deber a las proyecciones del DANE que no capturan del todo bien los flujos migratorios. Bogotá, aunque tiene buenos indicadores, no está en los primeros puestos, y departamentos como Guainía, Amazonas, Guaviare y Vaupés presentan las coberturas más bajas, lo cual pone en evidencia las brechas de acceso en regiones apartadas.
